```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\Abish\Downloads\archive\netflix_titles.csv")

print(df.head())
```

```python
df['type'].value_counts().plot(kind='bar')

plt.title("Movies vs TV Shows")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
df['rating'].value_counts().head(10).plot(kind='bar')

plt.title("Top Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
df['release_year'].value_counts().sort_index().plot(figsize=(10,5))

plt.title("Release Year Distribution")
plt.xlabel("Year")
plt.ylabel("Number of Shows")
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Countries")
plt.xlabel("Country")
plt.ylabel("Count")
plt.xticks(rotation=45)

plt.show()
```

```python
movies = df[df['type']=="Movie"]

movies['duration_num'] = movies['duration'].str.replace(" min","")

movies['duration_num'] = pd.to_numeric(movies['duration_num'], errors='coerce')

movies['duration_num'].dropna().hist(bins=20)

plt.title("Movie Duration")
plt.xlabel("Minutes")
plt.ylabel("Frequency")
plt.show()
```
